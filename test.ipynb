```python
import numpy as np
import matplotlib.pyplot as plt

W = np.zeros(2)
temp_W = np.zeros(2)

X = [4,3]
learn_rate = 0.01
min_w= np.full(shape=(2,1) , fill_value=float('inf'))

epoch = 50
loss = np.zeros(epoch)

all_w = np.zeros((epoch,2))

for i in range(epoch):
    
    y_model = np.dot(X,W)
    error = y_model - 12

    temp_W[0] = W[0] - learn_rate * error * X[0]
    temp_W[1] = W[1] - learn_rate * error * X[1]
    
    if (abs(temp_W[0] - W[0])) < 0.05 and ( (abs(temp_W[1] - W[1])) < 0.05) :
        break
    else:
        W[0] = temp_W[0]
        W[1] = temp_W[1]
        all_w[i] = [W[0] , W[1]]
        loss[i] = error ** 2
        print(W , "   " , i)
    
print("---------------------------")  
print(W)
print(min(loss))

print(len(all_w) , " - ", len(loss))
plt.scatter(all_w[:,0], loss)
plt.scatter(all_w[:,1] ,loss)

plt.show()
    
```

```python
from mpl_toolkits.mplot3d import Axes3D  

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(all_w[:, 0], all_w[:, 1], loss, c=loss, cmap='viridis')

ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Loss')

plt.show()
```
